```python
file_path = "/home/jovyan/Project2/data/poverty_rate.xlsx"

# Row numbers where each table's header starts
header_rows = [5, 58, 111, 164, 217, 270, 323, 376, 429, 482,
               535, 588, 641, 694, 747, 800, 853, 906, 959,
               1012, 1065, 1118, 1171, 1224, 1277, 1330, 1383,
               1436, 1489, 1542, 1595, 1648, 1701, 1754, 1807,
               1860, 1913, 1966, 2019, 2072, 2125, 2178, 2231,
               2284, 2337, 2390, 2443]

# Number of rows per table ( 53 states + 1 header)
nrows = 54

# Store tables
tables = []

for i, start_row in enumerate(header_rows):
    # Read each table
    df = pd.read_excel(file_path, skiprows=start_row - 1, nrows=nrows)
    
    # Skip if not a valid table
    if df.empty or "State" not in df.columns[0]:
        continue

    # Clean columns
    df.columns = [
        "State", "Total_population", "Number_in_poverty",
        "MOE_population", "Percent_in_poverty", "MOE_percent"
    ]

    # Get year (usually the row before the header)
    year_cell = pd.read_excel(file_path, skiprows=start_row - 3, nrows=1).iloc[0, 0]
    year_match = re.search(r'(\d{4})', str(year_cell))
    if year_match:
        year = int(year_match.group(1))
    else:
        continue


    df["Year"] = year
    tables.append(df)

# Combine all tables
poverty_all = pd.concat(tables, ignore_index=True)

# Filter to 19992020
poverty_filtered = poverty_all[(poverty_all["Year"] >= 1999) & (poverty_all["Year"] <= 2020)]

```

```python
poverty_filtered = poverty_filtered.copy()
poverty_filtered['Year'] = poverty_filtered['Year'].astype(str).str.extract(r'(\d{4})')[0].astype(int)


# filter for 19992020, these numbers are in thousands
poverty_filtered = poverty_filtered.query('1999 <= Year <= 2020')
poverty_filtered = poverty_filtered[~poverty_filtered['State'].str.match(r'^\d{4}')]
poverty_filtered.head()
```

```python
# poverty_filtered["Total_population"] = pd.to_numeric(
#     poverty_filtered["Total_population"], errors="coerce"
# )
# poverty_filtered["Number_in_poverty"] = pd.to_numeric(
#     poverty_filtered["Number_in_poverty"], errors="coerce"
# )
poverty_filtered["Percent_in_poverty"] = pd.to_numeric(poverty_filtered["Percent_in_poverty"], errors="coerce")
poverty_filtered.head(10)
```

```python
state_totals = (
    poverty_filtered.groupby("State")["Percent_in_poverty"]
    .mean()
    .reset_index(name="Poverty_percent")
)
```

## Explanation of Data

We got our mortality data from CDC Wonder, where we queried the Multiple Cause of Death (1999-2020) database. This means that cold or heat was not the underlying cause of death, but it was a contributing factor. 
For cold related deaths we used the following ICD10 codes:
- X31: Exposure to excessive natural cold
- T68: Hypothermia

For heat related deaths we used:
- X30: Exposure to excessive natural heat
- T67: Effects of heat and light

We then calculated mortality rate for each data set by: (# temperature related deaths / total population) * 100,000 for each state. 
Our weather data was queried from https://asmith-agdata.shinyapps.io/prism_downloader/_w_a6c5076cc76b4c3fb50f4b256b4bafb5/#tab-5484-1 which is a website to downlaod data from the PRISM weather service. This provide us with the minimum, average, and maximum temperature for each state. To get the number of cold days for a state we filtered the data frame for values less than 0 degrees celsius, or below 32 degrees farenheit, in the tmin column for each state. For hot days we filtered the tmax column for days greater than 32 degrees celsius, or above 89.4 degrees farenheit.
Poverty percentages were derived from a dataset from the US Census Bureau.

## Conclusion/Analysis
The first plot provides the strongest support for our hypothesis. It shows the difference between cold related and heat related death rates for each state. Negative values (red) indicate higher heat related mortality, while positive values (blue) indicate higher cold related mortality. Most states are shaded light to moderate blue, indicating higher cold related mortality. Alaska had the highest difference, shown by the dark blue color, which is not surprising. Arizona and Nevada, conversely, had higher heat related mortality rates. Overall, this plot suggests that cold related deaths are generally higher than heat-related deaths across the U.S.

The second plot, a simple bar chart comparing cold (blue) and heat (red) mortality rates by state, further supports our hypothesis. Except for Arizona and Nevada, cold related mortality rates are consistently higher.

The third plot compares the number of cold days (minimum temperature < 0C) and hot days (maximum temperature > 32C) for each state. This exploratory analysis indicates that cold days outnumber hot days in most states, which may help explain the higher cold related mortality rates.

The final two plots examine potential correlations between mortality rates and poverty percentage. Cold mortality showed a small negative correlation with poverty, while heat mortality showed a slightly stronger positive correlation. This could suggest that people in poverty have more access to warm infrastructure during cold periods, but less access to air conditioning during heat waves. This is speculative and would require further study to confirm.

For our spatial analysis, Morans I autocorrelation was calculated for the difference between mortality rates, as well as for cold related and heat related mortality rates separately. All Morans I values were positive with p-values < 0.05, indicating that states with similar mortality rates are geographically clustered rather than randomly distributed.

Finally, the total mortality rate from 1999 to 2020 was 39.72 for cold related deaths and 12.89 for heat related deaths. This confirms that, across the United States, cold related mortality rates are substantially higher than heat related mortality rates over this period.
